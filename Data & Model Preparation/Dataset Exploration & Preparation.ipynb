## Dataset for Visualization

```python
df_vis = df.copy()
```

```python
education_map = {1: 'High School', 2: 'Associate', 3: 'Bachelors', 4: 'Masters', 5: 'Doctorate'}
satisfaction_map = {1: 'Low', 2: 'Medium', 3: 'High', 4: 'Very High'}
performance_map = {1: 'Low', 2: 'Good', 3: 'Excellent', 4: 'Outstanding'}
balance_map = {1: 'Bad', 2: 'Fair', 3: 'Good', 4: 'Best'}
job_level_map = {1: 'Entry Level', 2: 'Mid Level', 3: 'Senior Level', 4: 'Manager', 5: 'Executive'}
stock_option_level_map = {0: 'No Stock', 1: 'Level 1', 2: 'Level 2', 3: 'Level 3'}
business_travel_map = {'Travel_Rarely': 'Rarely Travels', 'Travel_Frequently': 'Frequently Travels', 'Non-Travel': 'Never Travels'}
```

```python
df_vis['Education'] = df_vis['Education'].map(education_map)
df_vis['EnvironmentSatisfaction'] = df_vis['EnvironmentSatisfaction'].map(satisfaction_map)
df_vis['JobInvolvement'] = df_vis['JobInvolvement'].map(satisfaction_map)
df_vis['JobSatisfaction'] = df_vis['JobSatisfaction'].map(satisfaction_map)
df_vis['RelationshipSatisfaction'] = df_vis['RelationshipSatisfaction'].map(satisfaction_map)
df_vis['WorkLifeBalance'] = df_vis['WorkLifeBalance'].map(balance_map)
df_vis['JobLevel'] = df_vis['JobLevel'].map(job_level_map)
df_vis['StockOptionLevel'] = df_vis['StockOptionLevel'].map(stock_option_level_map)
df_vis['BusinessTravel'] = df_vis['BusinessTravel'].map(business_travel_map)
```

```python
df_vis['AgeGroup'] = pd.cut(df_vis['Age'], bins=[17, 25, 35, 45, 55, 65],
                             labels=['1825', '2635', '3645', '4655', '56-60'])

df_vis['IncomeGroup'] = pd.cut(df_vis['MonthlyIncome'], bins=[1000, 5000, 10000, 15000, 20000],
                                labels=['Low (1k-5k)', 'Medium (5k-10k)', 'High (10k-15k)', 'Very High (15k-20K)'])

df_vis['DistanceGroup'] = pd.cut(df_vis['DistanceFromHome'], bins=[0, 5, 10, 20, 30], 
                                 labels=['0-5', '6-10', '11-20', '21-30'])


df_vis['CompaniesWorkedGroup'] = pd.cut(df_vis['NumCompaniesWorked'], bins=[-1, 1, 3, 5, 10],
                                         labels=['01', '23', '45', '6+'])

df_vis['HikeCategory'] = pd.cut(df_vis['PercentSalaryHike'], bins=[10, 15, 20, 25],
                                 labels=['Low (1115%)', 'Medium (1620%)', 'High (2125%)'])

df_vis['ExperienceGroup'] = pd.cut(df_vis['TotalWorkingYears'], bins=[-1, 5, 10, 20, 30, 40],
                                    labels=['05', '610', '1120', '2130', '3140'])

df_vis['TrainingFreq'] = pd.cut(df_vis['TrainingTimesLastYear'], bins=[-1, 1, 3, 6],
                                 labels=['Low (01)', 'Medium (23)', 'High (46)'])

df_vis['TenureGroup'] = pd.cut(df_vis['YearsAtCompany'], bins=[-1, 2, 5, 10, 20, 40],
                                labels=['<3 yrs', '35 yrs', '610 yrs', '1120 yrs', '21+ yrs'])

df_vis['RoleTenure'] = pd.cut(df_vis['YearsInCurrentRole'], bins=[-1, 2, 5, 10, 15, 20],
                               labels=['<3 yrs', '35 yrs', '610 yrs', '1115 yrs', '1618 yrs'])

df_vis['PromotionRecency'] = pd.cut(df_vis['YearsSinceLastPromotion'], bins=[-1, 1, 3, 5, 10, 15],
                                     labels=['Recent (<1yr)', '13 yrs', '45 yrs', '610 yrs', '1115 yrs'])

df_vis['ManagerYearsGroup'] = pd.cut(df_vis['YearsWithCurrManager'], bins=[-1, 2, 5, 10, 15, 18],
                                      labels=['<3 yrs', '35 yrs', '610 yrs', '1115 yrs', '1618 yrs'])
```

```python
df_vis.rename(columns={
    'Attrition': 'Attrition Status',
    'BusinessTravel': 'Business Travel Frequency',
    'DistanceFromHome': 'Distance From Home (km)',
    'Education': 'Education Level',
    'EducationField': 'Education Field',
    'EnvironmentSatisfaction': 'Environment Satisfaction',
    'JobInvolvement': 'Job Involvement',
    'JobLevel': 'Job Level',
    'JobRole': 'Job Role',
    'JobSatisfaction': 'Job Satisfaction',
    'MaritalStatus': 'Marital Status',
    'MonthlyIncome': 'Monthly Income ($)',
    'NumCompaniesWorked': 'Count Previous Companies',
    'OverTime': 'Works Overtime',
    'PercentSalaryHike': 'Salary Hike (%)',
    'RelationshipSatisfaction': 'Relationship Satisfaction',
    'StockOptionLevel': 'Stock Option Level',
    'TotalWorkingYears': 'Total Working Years',
    'TrainingTimesLastYear': 'Training Times Last Year',
    'WorkLifeBalance': 'Work-Life Balance',
    'YearsAtCompany': 'Years at Company',
    'YearsInCurrentRole': 'Years in Current Role',
    'YearsSinceLastPromotion': 'Years Since Last Promotion',
    'YearsWithCurrManager': 'Years with Current Manager',
    'AgeGroup': 'Age Group',
    'IncomeGroup': 'Income Group',
    'DistanceGroup': 'Distance Group',
    'CompaniesWorkedGroup': 'Companies Worked Group',
    'HikeCategory': 'Salary Hike Category',
    'ExperienceGroup': 'Experience Group',
    'TrainingFreq': 'Training Frequency',
    'TenureGroup': 'Company Tenure Group',
    'RoleTenure': 'Role Tenure Group',
    'PromotionRecency': 'Promotion Recency Group',
    'ManagerYearsGroup': 'Years With Current Manager Group',
}, inplace=True)
```

## Dataset for Machine Learning

```python
df_ml = df.copy()
```

### Drop Features

```python
df_ml.drop(columns=['Department', 'EducationField', 'JobRole'],
        inplace=True)
```

### Feature Mapping

```python
df_ml = pd.get_dummies(df_ml, columns=['MaritalStatus', 'Gender'])
```

```python
mapping = {
    "BusinessTravel": {
        "Non-Travel": 0,
        "Travel_Rarely": 1,
        "Travel_Frequently": 2
    },

    "OverTime": {
        "No": 0,
        "Yes": 1
    },

    "Attrition": {
        "No": 0,
        "Yes": 1
    },
}
```

```python
df_ml.replace(mapping, inplace=True)
```

```python
df_ml.rename(columns={
    'Age': 'Age',
    'Attrition': 'Attrition Status',
    'BusinessTravel': 'Business Travel Frequency',
    'DistanceFromHome': 'Distance From Home (km)',
    'Education': 'Education Level',
    'EnvironmentSatisfaction': 'Environment Satisfaction',
    'Gender': 'Gender',
    'JobInvolvement': 'Job Involvement',
    'JobLevel': 'Job Level',
    'JobSatisfaction': 'Job Satisfaction',
    'MonthlyIncome': 'Monthly Income ($)',
    'NumCompaniesWorked': 'Number of Companies Worked',
    'OverTime': 'Works Overtime',
    'PercentSalaryHike': 'Salary Hike (%)',
    'RelationshipSatisfaction': 'Relationship Satisfaction',
    'StockOptionLevel': 'Stock Option Level',
    'TotalWorkingYears': 'Total Working Years',
    'TrainingTimesLastYear': 'Training Times Last Year',
    'WorkLifeBalance': 'Work-Life Balance',
    'YearsAtCompany': 'Years at Company',
    'YearsInCurrentRole': 'Years in Current Role',
    'YearsSinceLastPromotion': 'Years Since Last Promotion',
    'YearsWithCurrManager': 'Years with Current Manager',
    'Gender_Male': 'Gender: Male',
    'Gender_Female': 'Gender: Female',
    'MaritalStatus_Divorced': 'Marital Status: Divorced',
    'MaritalStatus_Married': 'Marital Status: Married',
    'MaritalStatus_Single': 'Marital Status: Single'
    }, inplace=True)
```
